```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
df = pd.read_csv('toronto_italian_pr3_restaurants_DONE.csv')
```

```python
# Assume 0 if missing
num_cols = ['positive_reviews', 'neutral_reviews', 'negative_reviews', 'OpenOnWeekends', 
            'NumberOfWeekdaysWithEarlyOpening', 'NumberOfWeekdaysWithLateClosing']
for col in num_cols:
    df[col] = df[col].replace(np.NaN, 0.0)
```

```python
# Assume false value if missing
cols = ['ambience_romantic', 'ambience_intimate', 'ambience_classy', 'ambience_hipster' ,'ambience_touristy',
       'ambience_trendy', 'ambience_upscale', 'ambience_casual', 'good_for_dessert', 'good_for_latenight',
       'good_for_lunch', 'good_for_dinner', 'good_for_breakfast', 'good_for_brunch', 'parking_garage',
       'parking_validated', 'parking_lot', 'parking_valet', 'bestnight_monday', 'bestnight_tuesday', 
       'bestnight_wednesday', 'bestnight_thursday', 'bestnight_friday', 'bestnight_saturday', 'bestnight_sunday',
       'music_dj', 'music_background_music', 'music_no_music', 'music_karaoke', 'music_live', 'music_video', 
       'music_jukebox', 'parking_street']
for col in cols:
    df[col] = df[col].replace(np.NaN, False)
```

```python
Y = df["stars"]
X = df[['ambience_romantic', 'ambience_intimate', 'ambience_classy', 'ambience_hipster', 'ambience_touristy', 'ambience_trendy', 
        'ambience_upscale', 'ambience_casual', 'good_for_dessert', 'good_for_latenight', 
        'good_for_lunch', 'good_for_dinner', 'good_for_breakfast', 'review_count', 'good_for_brunch','parking_garage', 
        'parking_street', 'parking_validated', 'parking_lot', 'parking_valet', 'bestnight_monday', 'full_bar',
        'beer_and_wine', 'bestnight_tuesday', 'bestnight_wednesday', 'bestnight_thursday','no_alcohol' ,'bestnight_friday', 'bestnight_sunday',
        'music_dj', 'music_background_music', 'music_karaoke', 'music_live', 'music_video', 'music_jukebox', 'smoking_outdoor',
        'smoking_no', 'smoking_yes', 'wifi_free', 'wifi_no', 'wifi_paid', 'noise_very_loud', 'noise_loud', 'noise_quiet',
        'noise_average', 'negative_reviews', 'neutral_reviews']]
X1 = sm.add_constant(X)
reg = sm.OLS(Y.astype(float), X1.astype(float)).fit()
reg.summary()
```
