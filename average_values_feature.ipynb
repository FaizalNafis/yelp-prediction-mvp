## Average Values of Features
Get averages of each feature column from data set. Dashboard can default to these values if no data is provided.

```python
import pandas as pd
```

```python
pd.set_option('display.max_colwidth', -1)
pd.set_option('display.max_columns', None)  
## Remove unneeded columns
df = pd.read_csv("datasets/odyssey_features_updated.csv") 
col_remove = ['address','categories','city','cuisine',
              'latitude','longitude','name','state',
              'cat','price_range_1','price_range_2',
              'price_range_3','price_range_4','postal_code',
              'neighborhood'
             ]
yelp_df = df.drop(col_remove, axis=1)
yelp_df.set_index('business_id', inplace=True)
yelp_df
```

```python
print('Monday hours average: ' + str(yelp_df['Monday'].mean()))
print('Tuesday hours average: ' + str(yelp_df['Tuesday'].mean()))
print('Wednesday hours average: ' + str(yelp_df['Wednesday'].mean()))
print('Thursday hours average: ' + str(yelp_df['Thursday'].mean()))
print('Friday hours average: ' + str(yelp_df['Friday'].mean()))
print('Saturday hours average: ' + str(yelp_df['Saturday'].mean()))
print('Sunday hours average: ' + str(yelp_df['Sunday'].mean()))
```

```python
print('TotalOpenTimeInWeek average: ' + str(yelp_df['TotalOpenTimeInWeek'].mean()))
print('NumberOfWeekdaysWithEarlyOpening average: ' + str(yelp_df['NumberOfWeekdaysWithEarlyOpening'].mean()))
print('NumberOfWeekdaysWithLateClosing average: ' + str(yelp_df['NumberOfWeekdaysWithLateClosing'].mean()))
print('OpenOnWeekends average: ' + str(yelp_df['OpenOnWeekends'].mean()))
```
