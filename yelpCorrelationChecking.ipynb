```python
# pd.options.display.max_rows = 9999
# pd.options.display.max_columns = 9999

df1 = df.drop(columns=['neighborhood', 'price_range_1', 'price_range_2', 'price_range_3', 'price_range_4',
                     'latitude', 'longitude', 'postal_code', 'music_no_music'])
c1 = df1.corr()
c1['good_for_dinner'] > 0.3
```

```python
corr_df = df1.corr()
updated = corr_df[['ambience_romantic', 'ambience_classy', 'ambience_hipster', 'ambience_trendy', 'ambience_casual', 
        'good_for_breakfast', 'review_count', 'good_for_brunch', 'parking_garage', 'parking_street', 'parking_lot', 
        'noise_quiet', 'negative_reviews']]
# updated [updated['parking_lot'] > 0.3]
df['stars'].mean()
```

```python
df2 = df1.drop(columns = ['stars'])
corr = df2.corr().abs()

s = corr.unstack()
so = s.sort_values(kind="quicksort")
so = so[so != 1.0]
s_1 = so[so <= 0.5]
s_2 = so[so > 0.5]

s_1 = pd.DataFrame({'vars':s_1.index, 'corr':s_1.values})

s_1[['var1','var2']] = pd.DataFrame(s_1.vars.values.tolist(), index= s_1.index)
s_1 = s_1.iloc[:,1:]

s_2 = pd.DataFrame({'vars':s_2.index, 'corr':s_2.values})

s_2[['var1','var2']] = pd.DataFrame(s_2.vars.values.tolist(), index= s_2.index)
s_2 = s_2.iloc[:,1:]

s_1.sort_values(by=['var1'])
```
