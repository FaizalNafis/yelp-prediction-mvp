```python
import pandas as pd
import numpy as np
import statsmodels.api as sm 
import statsmodels.formula.api as smf
```

```python
data = pd.read_csv(r"C:\Users\BigD\Desktop\UCL-Programming to BA\toronto_italian_pr2_restaurants_DONE.csv")
data.head()
```

```python
num_cols = ['positive_reviews', 'neutral_reviews', 'negative_reviews', 'OpenOnWeekends', 
            'NumberOfWeekdaysWithEarlyOpening', 'NumberOfWeekdaysWithLateClosing']
for col in num_cols:
    data[col] = data[col].replace(np.NaN, 0.0)
```

```python
cols = ['ambience_romantic', 'ambience_intimate', 'ambience_classy', 'ambience_hipster' ,'ambience_touristy',
       'ambience_trendy', 'ambience_upscale', 'ambience_casual', 'good_for_dessert', 'good_for_latenight',
       'good_for_lunch', 'good_for_dinner', 'good_for_breakfast', 'good_for_brunch', 'parking_garage',
       'parking_validated', 'parking_lot', 'parking_valet', 'bestnight_monday', 'bestnight_tuesday', 
       'bestnight_wednesday', 'bestnight_thursday', 'bestnight_friday', 'bestnight_saturday', 'bestnight_sunday',
       'music_dj', 'music_background_music', 'music_no_music', 'music_karaoke', 'music_live', 'music_video', 
       'music_jukebox', 'parking_street','no_alcohol']
for col in cols:
    data[col] = data[col].replace(np.NaN, False)
```

```python
Y = data["stars"]
X = data[['ambience_romantic', 'ambience_intimate', 'ambience_classy', 'ambience_hipster', 'ambience_touristy', 'ambience_trendy',
          'ambience_upscale', 'ambience_casual', 'good_for_dessert', 'good_for_latenight', 'good_for_lunch', 'good_for_dinner', 
          'good_for_breakfast', 'review_count', 'good_for_brunch', 'parking_garage', 'parking_street', 'parking_validated', 
          'parking_lot', 'parking_valet', 'bestnight_monday', 'full_bar', 'beer_and_wine', 'bestnight_tuesday', 'bestnight_wednesday', 
          'bestnight_thursday', 'bestnight_friday', 'bestnight_sunday', 'music_dj', 'music_background_music', 'music_karaoke', 'music_live', 
          'music_video', 'music_jukebox', 'smoking_outdoor', 'smoking_no', 'smoking_yes', 'wifi_free', 'wifi_no', 'wifi_paid', 'noise_very_loud', 
          'noise_loud', 'noise_quiet', 'noise_average', 'negative_reviews', 'neutral_reviews', 'positive_reviews', 'no_alcohol']]
X1 = sm.add_constant(X)
reg = sm.OLS(Y.astype(float), X1.astype(float)).fit()
reg.summary()
```

```python
X2 = X1.drop(['ambience_classy', 'good_for_dessert', 'parking_valet', 'bestnight_wednesday', 'bestnight_friday', 'negative_reviews', 'neutral_reviews','no_alcohol'], axis = 1)
reg = sm.OLS(Y.astype(float), X2.astype(float)).fit()
reg.summary()
```

```python
X3 = X2.drop(['ambience_upscale','parking_street','ambience_touristy','beer_and_wine','music_video','music_jukebox'], axis = 1)
reg = sm.OLS(Y.astype(float), X3.astype(float)).fit()
reg.summary()
```

```python
X4 = X3.drop(['ambience_intimate','good_for_brunch','parking_validated','music_live','smoking_no','wifi_no'], axis = 1)
reg = sm.OLS(Y.astype(float), X4.astype(float)).fit()
reg.summary()
```

```python
X5 = X4.drop(['ambience_trendy','smoking_yes','noise_average'], axis = 1)
reg = sm.OLS(Y.astype(float), X5.astype(float)).fit()
reg.summary()
```

```python
X6 = X5.drop(['noise_very_loud','noise_quiet'], axis = 1)
reg = sm.OLS(Y.astype(float), X6.astype(float)).fit()
reg.summary()
```

```python
X7 = X6.drop(['good_for_lunch','bestnight_tuesday','wifi_free'], axis = 1)
reg = sm.OLS(Y.astype(float), X7.astype(float)).fit()
reg.summary()
```

```python
X8 = X7.drop(['parking_garage','music_karaoke','wifi_paid'], axis = 1)
reg = sm.OLS(Y.astype(float), X8.astype(float)).fit()
reg.summary()
```

```python
X9 = X8.drop(['good_for_latenight','parking_lot'], axis = 1)
reg = sm.OLS(Y.astype(float), X9.astype(float)).fit()
reg.summary()
```
