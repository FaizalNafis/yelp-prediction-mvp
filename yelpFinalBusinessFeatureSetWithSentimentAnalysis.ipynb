```python
import pandas as pd
import numpy as np
from datetime import timedelta
```

```python
# merge restaurant business features with review sentiment analysis
reviews_sentiment_df = pd.read_csv("datasets/american_reviews_sentiment.csv")
restaurants_df = pd.read_csv('datasets/vegas_pr2_restaurants.csv')
restaurants_df['TotalOpenTimeInWeek'] = restaurants_df['TotalOpenTimeInWeek'].map(lambda x: pd.to_timedelta(x))
```

```python
#Rearrange reviews sentiment dataframe
reviews_df = reviews_sentiment_df.pivot(index='business_id', columns='label', values='size')
reviews_df = reviews_df.rename_axis(None, axis=1).reset_index() 
reviews_df.columns = ['business_id', 'negative_reviews', 'neutral_reviews', 'positive_reviews']
```

```python
# merge categorized reviews and business features
features_df = pd.merge(restaurants_df, reviews_df, how='left', on='business_id')
features_df
```

```python
len(restaurants_df)
```

```python
# Assume 0 if missing
num_cols = ['positive_reviews', 'neutral_reviews', 'negative_reviews', 'OpenOnWeekends', 
            'NumberOfWeekdaysWithEarlyOpening', 'NumberOfWeekdaysWithLateClosing']
for col in num_cols:
    features_df[col] = features_df[col].replace(np.NaN, 0.0)
```

```python
# Assume false value if missing
cols = ['ambience_romantic', 'ambience_intimate', 'ambience_classy', 'ambience_hipster' ,'ambience_touristy',
       'ambience_trendy', 'ambience_upscale', 'ambience_casual', 'good_for_dessert', 'good_for_latenight',
       'good_for_lunch', 'good_for_dinner', 'good_for_breakfast', 'good_for_brunch', 'parking_garage',
       'parking_validated', 'parking_lot', 'parking_valet', 'bestnight_monday', 'bestnight_tuesday', 
       'bestnight_wednesday', 'bestnight_thursday', 'bestnight_friday', 'bestnight_saturday', 'bestnight_sunday',
       'music_dj', 'music_background_music', 'music_no_music', 'music_karaoke', 'music_live', 'music_video', 
       'music_jukebox', 'parking_street']
for col in cols:
    features_df[col] = features_df[col].replace(np.NaN, False)
```

```python
features_df['TotalOpenTimeInWeek'] = features_df['TotalOpenTimeInWeek'].apply(lambda x: x.total_seconds())
mean = features_df['TotalOpenTimeInWeek'].mean()
features_df['TotalOpenTimeInWeek'] = features_df['TotalOpenTimeInWeek'].replace(np.NaN, mean)
```

```python
features_df.to_csv('datasets/vegas_sme_dataset.csv', mode='a', encoding='utf-8', index=False)
```
